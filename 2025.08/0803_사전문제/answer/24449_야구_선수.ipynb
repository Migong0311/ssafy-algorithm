
# 24449. 야구 선수

## 📝 문제 설명

N명의 야구 선수가 있고, 각 선수는 특정한 실력을 가지고 있다.

이 야구선수들은 팀을 정해야 하는데 팀을 정할때 규칙이 존재한다.

팀을 정할때 규칙은 **정해진 팀원 사이의 야구 실력의 차이가 최대 K를 넘지 않아야 한다**는 것이다.

야구 선수들의 실력 정보가 주어졌을 때, **팀의 인원이 최대한 많게 구성**하고 싶다.
최대가 되도록 구성했을 때, **몇 명이 되는지** 구하는 프로그램을 작성하시오.

---

## 📌 입력 형식

- 첫 번째 줄에 테스트케이스의 개수 T가 주어진다.
- 각 테스트케이스마다
  - 첫 줄에 `야구선수 인원 N`, `허용가능한 실력 K`가 주어진다.
  - 두 번째 줄에 N명의 야구선수들의 실력이 입력으로 주어진다.

---

## 📌 출력 형식

- 각 테스트케이스마다 `#테스트케이스번호 최대팀원수` 형식으로 출력한다.

---

## 📚 입력 예시

```
3
4 2
6 4 2 3
4 3
1 2 3 4
4 1
1 3 7 5
```

---

## ✅ 출력 예시

```
#1 3
#2 4
#3 1
```
    

````

T = int(input())  # 테스트 케이스 개수 입력

for t in range(1, T + 1):
    N, K = map(int, input().split())  # 선수 수 N, 실력 차이 최대치 K
    skills = list(map(int, input().split()))  # 선수들의 실력 리스트

    skills.sort()  # 실력을 오름차순 정렬

    max_team = 0  # 최대 팀원 수 초기화

    for i in range(N):
        for j in range(i, N):
            if skills[j] - skills[i] <= K:  # 실력 차이가 K 이하인 경우
                team_size = j - i + 1
                if team_size > max_team:
                    max_team = team_size  # 최대 팀원 수 갱신
            else:
                break  # K를 초과하면 더 이상 볼 필요 없음

    print(f"#{t} {max_team}")  # 결과 출력
    
````
