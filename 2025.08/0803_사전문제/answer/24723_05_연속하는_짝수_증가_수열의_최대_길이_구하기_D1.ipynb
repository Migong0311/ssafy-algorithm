# 24723. 05 연속하는 짝수 증가 수열의 최대 길이 구하기
**Level: D1**


N개의 양의 정수로 이루어진 수열이 주어질 때, 그 수열 안에는 **연속하며 증가하는 짝수**만으로 이루어진 부분 수열이 존재합니다.

예를 들어 주어진 수열이 `2 4 6 8 10` 이라면  
- 이 안에 연속하며 증가하는 짝수 부분 수열은 `2 4 6 8 10`이고, 길이는 5입니다.

또는 수열이 `2 2 2 2 2` 라면  
- 모든 수는 짝수이지만 증가하지 않기 때문에, 최대 길이는 1입니다.

홀수만 있는 경우는 연속 증가 짝수 수열이 없으므로 길이는 0이 됩니다.

---

## 입력 형식
```
첫 줄에 테스트 케이스의 개수 T가 주어집니다.  
이후 각 테스트 케이스마다,  
첫 줄에 정수의 개수 N,  
둘째 줄에 N개의 양의 정수가 주어집니다.
```

## 출력 형식
```
각 테스트케이스마다 #<테스트번호> <가장 긴 길이> 를 출력합니다.
```

## 예제
### 입력
```
3  
5  
2 4 6 8 10  
5  
2 2 2 2 2  
5  
1 3 5 7 9
```

### 출력
```
#1 5  
#2 1  
#3 0
```


````
T = int(input())  # 테스트 케이스 개수를 입력받음

for test_case in range(1, T + 1):  # 1부터 T까지 반복
    N = int(input())  # 수열의 길이 입력
    sequence = list(map(int, input().split()))  # 수열을 리스트로 입력받음

    longest_length = 0  # 최대 길이를 저장할 변수
    current_streak = 0  # 현재 연속 증가하는 짝수 수열의 길이
    previous_number = -1  # 이전 숫자를 저장할 변수

    for current_number in sequence:
        if current_number % 2 == 0:  # 짝수일 경우
            if current_streak == 0 or current_number > previous_number:
                current_streak += 1  # 증가하는 짝수라면 현재 길이 +1
            else:
                current_streak = 1  # 증가하지 않으면 다시 시작
            previous_number = current_number  # 이전 숫자 업데이트
        else:
            current_streak = 0  # 홀수면 연속 종료
            previous_number = -1  # 초기화

        longest_length = max(longest_length, current_streak)  # 최대 길이 갱신

    print(f'#{test_case} {longest_length}')  # 출력 형식에 맞춰 출력
````
