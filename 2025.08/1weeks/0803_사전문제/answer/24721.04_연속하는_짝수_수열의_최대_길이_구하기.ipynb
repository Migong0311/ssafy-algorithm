# 24721.04 - 연속하는 짝수 수열의 최대 길이 구하기

## 📝 문제 설명

N개의 양의 정수로 이루어진 수열이 주어질 때, 그 수열 안에는 연속하는 **짝수만으로 이루어진 부분 수열**이 존재한다.

예를 들어 수열이 `9 8 6 8 2` 라면 연속한 짝수 부분 수열로는 `8 6`, `8 6 8`, `8 6 8 2` 등이 있고, 이 중에서 가장 긴 수열은 `8 6 8 2` 이며 길이는 `4`이다.

### ✅ 입력 형식
- 첫 줄에 테스트케이스의 개수 T가 입력된다.
- 각 테스트 케이스의 첫 줄에는 수열의 길이 N이 주어지고,
- 두 번째 줄에는 N개의 양의 정수가 공백으로 구분되어 주어진다.

### ✅ 출력 형식
- 각 테스트케이스마다 `#{테스트케이스번호} 최대길이` 형식으로 출력한다.

### 💡 예제 입력
```
3
5
9 8 6 8 2
5
6 2 2 4 10
5
1 3 5 7 9
```

### 💡 예제 출력
```
#1 4
#2 5
#3 0
```


````
T = int(input())  # 테스트 케이스 수 입력받음

for t in range(1, T + 1):  # 테스트케이스 반복
    N = int(input())  # 수열의 길이 입력받음
    seq = list(map(int, input().split()))  # 수열 입력받음

    max_len = 0  # 최대 길이 저장용
    current_len = 0  # 현재 짝수 연속 길이 저장용

    for num in seq:
        if num % 2 == 0:  # 짝수일 경우
            current_len += 1  # 현재 길이 증가
            max_len = max(max_len, current_len)  # 최대값 갱신
        else:
            current_len = 0  # 홀수 만나면 초기화

    print(f"#{t} {max_len}")  # 출력 형식에 맞게 출력
````
