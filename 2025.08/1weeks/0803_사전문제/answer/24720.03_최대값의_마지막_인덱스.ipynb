# 24720.03 최대값의 마지막 인덱스

## 📝 문제 설명

숫자 N개가 입력으로 주어 질 때, 그 중에 가장 큰 값과 그 값이 **마지막으로 등장하는 인덱스**를 출력하는 프로그램을 작성하세요.

- 첫 줄에 테스트 케이스 개수 T 가 입력으로 주어집니다.
- 두 번째 줄부터 각 테스트케이스의 입력이 주어집니다.
- 각 테스트케이스는 두 줄로 구성되며:
  - 첫 번째 줄은 숫자의 개수 N
  - 두 번째 줄은 N개의 숫자입니다.

### ✅ 출력 형식
```
#테스트케이스번호 최대값 인덱스
```

단, 최대값이 여러 번 등장하면 **가장 마지막에 등장한 인덱스**를 출력해야 합니다.


### 📌 입력 예시
```
3
5
100 1 2 3 100
5
99 99 1 2 3
5
4 3 2 4 1
```

### ✅ 출력 예시
```
#1 100 4
#2 99 1
#3 4 3
```


## 🔍 풀이 전략

1. 테스트 케이스 수 `T`를 입력받는다.
2. 각 테스트 케이스마다:
   - 숫자 개수 `N`을 입력받고
   - 숫자 리스트를 입력받는다.
3. 리스트에서 `최댓값(max)`을 구한다.
4. 리스트를 뒤집어서(`[::-1]`) **최댓값이 마지막으로 등장한 위치**를 찾아,
   이를 원래 리스트 기준 인덱스로 변환한다.
5. 출력은 형식에 맞춰 테스트케이스 번호, 최대값, 인덱스를 출력한다.


````
T = int(input())  # 테스트 케이스 개수를 입력받음

for t in range(1, T + 1):  # 1부터 T까지 반복함
    # 이번 테스트 케이스에서 숫자의 개수를 입력받음
    N = int(input())
    # 숫자들을 정수 리스트로 변환함 
    lst = list(map(int, input().split()))

    # 숫자 리스트에서 최댓값 찾기
    max_number = max(lst)
    # 리스트를 뒤집어서 최댓값이 마지막으로 등장한 위치를 구한 뒤, 원래 리스트에서의 인덱스로 변환함
    last_index = len(lst) - 1 - lst[::-1].index(max_number)

    print(f'#{t} {max_number} {last_index}')  # 반복문 안에 출력 넣어줬음
````
